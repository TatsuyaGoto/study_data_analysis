# Notebook(google colabo)の使い方
### Notebook(google colabo)とは
ブラウザ上で動作する対話型実行環境 統計のモデリングや機械学習などデータ分析に使用されることが想定されており、
データの視覚化などの作業に適しています。対話型の開発環境であるため、前の実行結果に応じて、
次に実行するプログラムや作業を選択できる。なお、実行した結果は作業履歴として記録に残せる。

1セルに纏めて記述

```python
# j:入力して上部にある ▶run でセル実行
# g:セルの左にある▶でセル実行
a = 5
b = 2
c = a * b
print(c)
```

分けて記述

```python
# 1行ずつ書いてもいい
a = 3
```

```python
# shift + enter でセル実行
b = 4
```

```python
# 結果を出力した場合は下に表示される
print(a*b)
```

セルの実行順番による違い

```python
print(c)
```

```python
# セルを実行してから上のセルを再度実行する
# 上から下ではなく、実行された順番
c = 5
```
